# On tossing coins

Suppose a fair coin is tossed for five times. What is the probability of observing five heads? ... (some brief explanation)

Consider what happens if the experiment is repeated with $N=100$ coins ...

```python
import matplotlib.pyplot as plt

import numpy as np
from numpy.random import randint

import seaborn as sb

"""Flip 100 coins for 5 times. Calculate the number of heads for each
coin."""
coin_tosses = randint(low=0, high=2, size=(5, 100))

num_heads = np.reshape(np.sum(coin_tosses, axis=0), (5, 20))

"""Visualize the data."""
fig = plt.figure(figsize=(10, 2), dpi=160)

ax = fig.add_subplot(111)
sb.heatmap(num_heads, annot=True, fmt='d', linewidth=0.1,
           cmap='Blues', ax=ax)
fig.subplots_adjust(left=None, bottom=0.2, top=0.9)
plt.show()
```

Now the conclusion that we can draw from this experiment is that when _multiple_ coins are tossed at the same time, the probability of observing five heads _increases_ dramatically. This why you would be very concerned about a study that (say) compared the blood-oxygen-level dependent (BOLD) signal between two conditions in 100,000 voxels and did not correct the results for multiple testing. Indeed, Bennett and co-authors have observed that ... (post-mortem salmon in fMRI paper)
